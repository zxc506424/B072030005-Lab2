```python
import pygame
import time

WIN_WIDTH = 1024
WIN_HEIGHT = 600
BTN_WIDTH = 80
BTN_HEIGHT = 80
HP_WIDTH = 40
HP_HEIGHT = 40
ENEMY_WIDTH = 50
ENEMY_HEIGHT =50
T_WIDTH =10
T_HEIGHT = 10

FPS = 30

# color (RGB)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# initialization
pygame.init()
# load image (background, enemy, buttons)
background_image = pygame.transform.scale(pygame.image.load("images/Map.png"), (WIN_WIDTH, WIN_HEIGHT))
enemy_image = pygame.transform.scale(pygame.image.load("images/enemy.png"), (50, 50))
muse_button_image = pygame.transform.scale(pygame.image.load("images/muse.png"), (BTN_WIDTH, BTN_HEIGHT))
sound_button_image = pygame.transform.scale(pygame.image.load("images/sound.png"), (BTN_WIDTH, BTN_HEIGHT))
continue_button_image = pygame.transform.scale(pygame.image.load("images/continue.png"), (BTN_WIDTH, BTN_HEIGHT))
pause_button_image = pygame.transform.scale(pygame.image.load("images/pause.png"), (BTN_WIDTH, BTN_HEIGHT))
hp_image = pygame.transform.scale(pygame.image.load("images/hp.png"), (HP_WIDTH, HP_HEIGHT))
dropped_hp_image = pygame.transform.scale(pygame.image.load("images/hp_gray.png"), (HP_WIDTH, HP_HEIGHT))
# ...(to be done)

# set the title
# ... (to be done)
class Game:
    def __init__(self):
        # window
        # ...(to be done)
        self.win = pygame.display.set_mode((WIN_WIDTH, WIN_HEIGHT))
        pygame.display.set_caption("My first game")
        # hp
        self.hp = 7
        self.max_hp = 10
        pass

    def game_run(self):
        # game loop
        run = True
        clock = pygame.time.Clock()
        while run:
            # event loop
            clock.tick(FPS)
            # event loop (user action)
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    run = False
                    
            # draw background
            # ...(to be done)
            self.win.blit(background_image, (0, 1))
            # draw enemy and health bar
            # ...(to be done)
            self.win.blit(enemy_image, (1, 280))
            pygame.draw.rect(self.win, (255, 0, 0), [1, 260, 70, 10])
            
            # draw menu (and buttons)
            # (... to be done)
            pygame.draw.rect(self.win, (0, 0, 0), [0, 0, 1024, 80])
            self.win.blit(pause_button_image, (930, 0))
            self.win.blit(continue_button_image, (850, 0))
            self.win.blit(sound_button_image, (770, 0))
            self.win.blit(muse_button_image, (690, 0))
            # draw HP
            self.win.blit(hp_image, (420, 8))
            self.win.blit(hp_image, (460, 8))
            self.win.blit(hp_image, (500, 8))
            self.win.blit(hp_image, (540, 8))
            self.win.blit(hp_image, (580, 8))
            self.win.blit(hp_image, (420, 45))
            self.win.blit(hp_image, (460, 45))
            self.win.blit(dropped_hp_image, (500, 45))
            self.win.blit(dropped_hp_image, (540, 45))
            self.win.blit(dropped_hp_image, (580, 45))
            
            pygame.display.update()

        pygame.quit()

if __name__ == "__main__":
    covid_game = Game()
    covid_game.game_run()
```
